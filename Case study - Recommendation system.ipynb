```python
#Task no 3 creating matrix for ratings dataset with userid and movieid and ratings as values
```

```python
Ratings = ratings.pivot(index='userId',columns= 'movieId',values='rating').fillna(0)
Ratings
```

```python
#task no 3 creating svd dataset with ratings dataset and computing three fold validation
```

```python
from surprise import Dataset, Reader, SVD
from surprise.model_selection import cross_validate
reader = Reader()
data = Dataset.load_from_df(ratings[['userId','movieId','rating']], reader)
svd = SVD()
cross_validate(svd,data,measures = ['RMSE', 'MAE'], cv = 3, verbose=True)
```

```python
#task no 5 printing 5 rated movies by the user id 5
```

```python
ratings_1 =ratings[(ratings.userId == 5) & (ratings.rating== 5)]
index=range(1,39)
ratings_1['S#']=index
ratings_1=ratings_1[['movieId','S#']].set_index('S#')
ratings_1=ratings_1.join(movies.title)
ratings_1
```

```python
#task no 5 swallow copying movies dataset and predicting top 10 recommodations for user id 5  
```

```python
user_5 = movies.copy()
trainset= data.build_full_trainset()
svd.fit(trainset)
ratings_1['Estimate_Score']=ratings_1['movieId'].apply(lambda x: svd.predict(1,x).est)
userid_5=ratings_1.sort_values('Estimate_Score',ascending= False)
top_10_user_5= userid_5.head(10)
```
